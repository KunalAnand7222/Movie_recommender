### What is a Movie Recommender System??
A movie recommender system is an application that suggests films to users based on various criteria, such as their preferences, viewing history, or the preferences of similar users. These systems are a key component of streaming platforms and online movie servicescy.racy.

### Components of a Recommender System:
-  Data Collection: Gathering user data such as ratings, viewing history, and explicit feedback.
- Data Processing: Cleaning and organizing data to make it suitable for analysis.
-  Model Building: Developing algorithms to analyze data and generate recommendations.
-  Recommendation Generation: Providing personalized movie suggestions to use

### Algorithms Used in Recommender Systems-  
Nearest Neighbors: Identifies users or items that are simila to each otherr
-  Matrix Factorization: Decomposes the user-item interaction matrix into latent factors.

### Benefits of Recommender Systems-  
Personalization: Offers a tailored viewing experience for each user
-  Increased Engagement: Keeps users engaged by suggesting relevant content.
-  Discovery: Helps users discover new movies they might not find otherwise..se..ta.rs.rs..:.rs. users.

```python
# fetching name value from genres column 
def name(obj):
    name1=[]
    for i in ast.literal_eval(obj):
        name1.append(i["name"])
    return name1
            
```

```python
# Updating cast column on the basis of character
# fetching top 5 characters from each row
def name1(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 5:
            L.append(i['name'])
        counter+=1
    return L 
```

```python
movie1['cast'] = movie1['cast'].apply(name1)
movie1.head()
```

```python
# fetching the director of the movie from the crew column
def director(obj):
    lis=[]
    for i in ast.literal_eval(obj):
        if i["job"]=="Director":
            lis.append(i["name"])
    return lis
```

```python
# updating it to movie1 
movie1["crew"]=movie1["crew"].apply(director)
```

```python
# replacing space with no gap and concatenate all
# You can use apply() on a pandas Series to apply a function to each element of the Series.
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movie1['cast'] = movie1['cast'].apply(collapse)
movie1['crew'] = movie1['crew'].apply(collapse)
movie1['genres'] = movie1['genres'].apply(collapse)
movie1['keywords'] = movie1['keywords'].apply(collapse)
```

```python
# converting overview column into list
movie1["overview"]=movie1["overview"].apply(lambda x:x.split())
```

```python
# concatenating all and making a new column
movie1["tags"]=movie1["overview"]+movie1["genres"]+movie1["cast"]+movie1["crew"]
```

```python
# Now if we want to recommend movies, we have to first find similarity between each movie
# one way is to do is find similar word from all movie tags and compare it but its too lengthy
# Another way is using vectorization: we have to convert all movie tags into vector 
# and after finding all vector, the one which is closest vector is to be recommended
# convert text=> vector(bag of words)
# combining all tags and find most 5000 or 8000 most repeated words 
# and then compare it with all movie tags how many times it repeated and then making an array
# when we are fetching most common words we are not going to consider stopwords
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=8000,stop_words='english')
# it will fetch most common 8000 words without considering stopwords
```

```python
count_vector=cv.fit_transform(movie1["tags"]).toarray()
# fit transform return sparse matrix(number of zero elements is significantly higher than the number of non-zero elements.) so we should convert into array
```

```python
# now replacing most words to its common names like dancing,dance,to danc
# natural language processing library
import nltk
```

```python

from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def common(obj):
    l=[]
    for i in obj.split():
        l.append(ps.stem(i))
    return " ".join(l)
# split=> conv string to list
# join => list to string
```

```python
# it will give common words for given words
ps.stem("dance")
```

```python
#converting similar words to common one and updating it to tags columns
movie1["tags"]=movie1["tags"].apply(common)
```

```python
# fetching again most common 8000 words but this time not the similar one
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=8000,stop_words='english')
count_vector=cv.fit_transform(movie1["tags"]).toarray()
cv.get_feature_names_out()
```

```python
# now we have to find distance between all the vector (not euclaridian distance we are going to find cosine value)
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# cosine value lies between 0 and 1 ...if 1 similarity is most...cosine value is directly proportional to similarity
similarity=cosine_similarity(count_vector)
# if we calculate similarity score between same movie it will be 1
```

```python
# shape of similarity
similarity.shape
```

```python
# first find index of given movie
# then use enumerate to hold similarity value along with index then sort in desecnding order
# then print top 10 movies names
```

```python
# finding top 5 best recommended moviex
def suggest(movie):
    indexx=movie1[movie1["title"]==movie].index[0]
    distance=sorted(list(enumerate(similarity[indexx])),reverse=True,key = lambda x: x[1])
    for i in distance[1:11]:
        print(movie1.iloc[i[0]].title)
        
```

```python
# converting all movie name to lower case and update it 
movie1["title"]=movie1["title"].apply(lambda x:x.lower())
```

```python
# printing top 10 suggested movie 
suggest("superman")
```
